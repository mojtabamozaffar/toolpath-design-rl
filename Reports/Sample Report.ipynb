```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True, font_scale=1)
```

```python
files = ['Tensorboard_data/max_moves_400_1/Total reward.csv',
         'Tensorboard_data/max_moves_400_2/Total reward.csv',
         'Tensorboard_data/max_moves_400_3/Total reward.csv',
        ]
```

```python
max_length = 100
n_tests = 1
data = np.array([np.loadtxt(file, delimiter=',', skiprows=1)[:max_length,2] for file in files])
data_mean = np.mean(data, axis=0)
data_std = np.std(data, axis=0)
```

```python
fig, ax = plt.subplots()
clrs = sns.color_palette(n_colors=n_tests)

epochs = list(range(max_length))
for i in range(n_tests):
    ax.plot(epochs, data_mean, label='max steps: 400', c=clrs[i])
    ax.fill_between(epochs, data_mean-data_std, data_mean+data_std ,alpha=0.3, facecolor=clrs[i])
ax.set_xlabel('Training iteration', color='k')  
ax.set_ylabel('score', color='k') 
ax.legend()
plt.show()

# fig.savefig('dense.png', format='png', dpi=600)
```
